# Librerías en python

## urllib3
Lectura de un archivo online

```python
import urllib3

## la URL del archivo
url= "http://winterolympicsmedals.com/medals.csv"
http=urllib3.PoolManager()
resp=http.request('GET', url)
print("conexión exitosa")
```

### codificación UTF-8 es internacional

```python
## conversión a la codificación UTF-8

datos= resp.data.decode('UTF-8')
print(type(datos))
print(datos)
```

### Procesamiento de los datos

```python
## lista con todos los datos del archivo
listaDatos=datos.split()
len(listaDatos)
```

```python
## lista de las lineas del archivo, cada linea es un string
listaLineas=datos.split('\n')
```

```python
listaLineas[0]
```

```python
type(listaLineas[0])
```

```python
## lista de listas
## 8 cols
listaL=list()
for li in listaLineas:
    #print(li)
    listaS=li.split(',')
    #print(listaS)
    #print(dato)
    listaL.append(listaS)

print(listaL)
```

```python
listaL[0][2]
```

```python
cols=listaL[0]
print("Las columnas del archivo medals.csv son: ",cols)
```

```python
long=len(listaL)
#print(long)
años=list()

for i in range(1,len(listaL)):
    años.append(listaL[i][0])
    #print(listaL[i][0])
#print(años)

conjuntoA=set(años)
print(conjuntoA)
lista_años= list(conjuntoA)
print(lista_años)
```

```python
ciudades=list()

for i in range(1, len(listaL)):
    ciudades.append(listaL[i][1])
lista_ciudades= list(set(ciudades))
print(lista_ciudades)
```

```python
deportes=list()
for i in range(1, len(listaL)):
    deportes.append(listaL[i][2])
lista_deportes= list(set(deportes))
print(lista_deportes)
```

```python
## National Olympic Comitee
NOC= list()
for i in range(1, len(listaL)):
    NOC.append(listaL[i][4])
lista_NOC= list(set(NOC))
print(lista_NOC)
```

### Creando un diccionario con los datos del archivo

```python
## dict vacio
dic_medals={}
for c in cols:
    dic_medals[c]=[]
```

```python
## llenar el diccionario con los datos
for l in listaL[1:]:
    #print(l)
    for i in range(len(cols)):
        dic_medals[cols[i]].append(l[i])
```
