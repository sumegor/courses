```python
def leer(f_name):
    print("Leyendo archivo...")
    f=open(f_name, 'r')
    texto=f.read()
    f.close()
    print('Fin de lectura del archivo')
    return texto
```

```python
archivoCompleto=ruta_abs+"medallas.csv"
cont=leer(archivoCompleto)
print(cont)
```

## Escribir archivos

```python
def escribir(f_name, permiso, texto_output):
    print("Inicio de escritura...")
    f=open(f_name, permiso)
    f.write(texto_output)
    f.close()
    print("Escritura finalizada!")    
```

### Crear el archivo y escribir en el
permiso = x

```python
escribir("archivoNuevo.txt",'x',"La linea 1 del archivo")
```

### El archivo ya existe y agrega al final texto
permiso = a (append)

```python
escribir("archivoNuevo.txt", 'a', "\n Esta es la linea 2 del archivo nuevo \n Esta es la linea 3 del archivo")
```
