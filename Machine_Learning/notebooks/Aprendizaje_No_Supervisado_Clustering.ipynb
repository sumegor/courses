## Aprendizaje No Supervisado - Descubriendo Patrones Ocultos

**Introducción, Clustering y Métodos Avanzados**
El Aprendizaje No Supervisado es una rama del aprendizaje automático (Machine Learning) que se enfoca en analizar y descubrir patrones ocultos en conjuntos de datos que no están etiquetados.

En esencia, el algoritmo de aprendizaje no supervisado recibe datos de entrada sin ninguna guía predefinida sobre la salida correcta. Su tarea es encontrar estructuras inherentes, relaciones, similitudes y agrupaciones dentro de esos datos por sí solo.

**Puntos clave sobre el Aprendizaje No Supervisado:**

* Datos sin etiquetas: La característica distintiva es la ausencia de una variable objetivo o etiquetas preasignadas que indiquen la "respuesta correcta" para cada dato.

* Descubrimiento de patrones: El objetivo principal es identificar patrones que no son evidentes a simple vista. Esto puede incluir la formación de grupos (clustering), la reducción de la dimensionalidad de los datos, o la identificación de asociaciones entre variables.

* Exploratorio: A menudo se utiliza como una técnica exploratoria para obtener información sobre la estructura subyacente de los datos antes de aplicar otras técnicas de aprendizaje automático.

* Flexibilidad: Los algoritmos no supervisados son más flexibles ya que no están restringidos por etiquetas predefinidas, lo que les permite descubrir patrones inesperados.

* Evaluación desafiante: Evaluar el rendimiento de los algoritmos no supervisados puede ser más difícil que en el aprendizaje supervisado, ya que no hay una "verdad fundamental" directa con la cual comparar las salidas. Se utilizan métricas basadas en la estructura de los datos encontrados.


### Introducción al Aprendizaje No Supervisado

**Definición:** Aprendizaje automático donde el algoritmo aprende patrones directamente de los datos **sin etiquetas o variables objetivo predefinidas**.

El objetivo es **descubrir la estructura inherente, las relaciones y las agrupaciones** dentro de los datos.

### Introducción a Clustering: Agrupando Datos Similares

* **Definición:** El clustering es una técnica de aprendizaje no supervisado que busca **agrupar automáticamente** puntos de datos similares en "clusters" o grupos.

* La similitud se define típicamente por la **distancia** entre los puntos de datos en el espacio de características.

* El objetivo es que los puntos dentro de un mismo cluster sean **más similares entre sí** que los puntos pertenecientes a otros clusters.

* **Aplicaciones:** Segmentación de clientes, análisis de imágenes, detección de comunidades en redes sociales, etc.

### Diapositiva 6: Clustering con K-Means - Teoría

* **Algoritmo iterativo** que busca particionar los datos en *k* clusters, donde *k* es un número predefinido por el usuario.

* **Pasos principales:**
    * **Inicialización:** Seleccionar aleatoriamente *k* centroides iniciales.
    * **Asignación:** Asignar cada punto de datos al cluster cuyo centroide esté más cercano (usualmente usando distancia euclidiana).
    * **Actualización:** Recalcular la posición de cada centroide como la media de todos los puntos asignados a ese cluster.
    * **Iteración:** Repetir los pasos de asignación y actualización hasta que los centroides converjan (cambien muy poco) o se alcance un número máximo de iteraciones.

* **Objetivo:** Minimizar la **inercia** (Within-Cluster Sum of Squares - WCSS), que es la suma de las distancias cuadradas de cada punto a su centroide asignado.

### Diapositiva 7: Aplicación Práctica de K-Means

**Ejemplo:** Segmentación de clientes basado en dos características (ej. gasto anual y frecuencia de compra).

**Visualización:**
* Un gráfico de dispersión de los puntos de datos con los dos ejes representando las características.
* Mostrar el proceso iterativo de K-Means: centroides iniciales, asignación de puntos a los centroides más cercanos (quizás con colores diferentes para cada cluster), y el movimiento de los centroides en cada iteración hasta la convergencia.
* Finalmente, mostrar los clusters formados con los centroides finales marcados.

**Código de Ejemplo (Python con scikit-learn):**

```
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (6 pares de datos)
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])

# Inicializar K-Means con k=2
kmeans = KMeans(n_clusters=2, random_state=0, n_init='auto')

# Ajustar el modelo a los datos
kmeans.fit(X)

# Obtener las etiquetas de los clusters y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red')
plt.title('Clustering K-Means')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

```
# prompt: * Un gráfico de dispersión de los puntos de datos con los dos ejes representando las características.
# * Mostrar el proceso iterativo de K-Means: centroides iniciales, asignación de puntos a los centroides más cercanos (quizás con colores diferentes para cada cluster), y el movimiento de los centroides en cada iteración hasta la convergencia.
# * Finalmente, mostrar los clusters formados con los centroides finales marcados.
# Utilizar 100 pares de datos (x,y) generados aleatoriamente para este ejemplo

import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Generar 100 pares de datos (x, y) aleatorios
np.random.seed(0)
X = np.random.rand(100, 2)

# Número de clusters
k = 3

# Inicializar K-Means
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)

# Ajustar el modelo a los datos
kmeans.fit(X)

# Obtener las etiquetas de los clusters y las coordenadas de los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizar los resultados
plt.figure(figsize=(10, 6))

# Mostrar los puntos de datos con colores según su cluster
for i in range(k):
  plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i+1}')

# Mostrar los centroides
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='*', c='black', label='Centroides')

plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Clustering con K-Means')
plt.legend()
plt.show()
```

### Evaluación de la Calidad del Clustering


* A diferencia del aprendizaje supervisado, la evaluación es más desafiante ya que no hay etiquetas "verdaderas" para comparar.

* **Métricas internas (sin usar información externa):**
    * **Inercia (Within-Cluster Sum of Squares - WCSS):** Mide qué tan compactos son los clusters. Un valor bajo es deseable, pero disminuye a medida que *k* aumenta. Se utiliza el **método del codo** para ayudar a elegir un *k* óptimo. (Mostrar un gráfico del método del codo).
    * **Coeficiente de Silueta:** Mide qué tan bien cada punto se ajusta a su cluster en comparación con otros clusters. El valor oscila entre -1 y 1. Valores cercanos a 1 indican clusters bien separados. (Mostrar la fórmula y la interpretación).

* **Métricas externas (usando información externa si está disponible, principalmente para comparar algoritmos):**
    * **Índice de Rand Ajustado (ARI):** Mide la similitud entre las agrupaciones obtenidas por el algoritmo y una clasificación externa conocida.
    * **Información Mutua Normalizada (NMI):** Similar al ARI, mide la similitud entre las agrupaciones basadas en la teoría de la información.

**Las métricas internas son más comunes en el aprendizaje no supervisado puro.**

### Método del Codo para K-Means: Eligiendo el Número Óptimo de Clusters (k)

**Explicación:**
* Se ejecuta K-Means para un rango de valores de *k* (ej. de 1 a 10).
* Se calcula la inercia para cada valor de *k*.
* Se grafica la inercia en función de *k*.
* El "codo" en la gráfica (el punto donde la disminución de la inercia se vuelve mucho más lenta) se considera un buen indicador del número óptimo de clusters.

**Gráfico:** Un ejemplo de un gráfico del método del codo mostrando una disminución pronunciada de la inercia para valores pequeños de *k* y una disminución más gradual después del "codo".

---

```
# prompt: Un ejemplo de un gráfico del método del codo mostrando una disminución pronunciada de la inercia para valores pequeños de *k* y una disminución más gradual después del "codo".

import matplotlib.pyplot as plt

# Datos de ejemplo para la inercia (reemplazar con datos reales calculados por KMeans)
k_values = range(1, 11)  # Valores de k de 1 a 10
inertia_values = [100, 80, 60, 50, 40, 35, 32, 30, 28, 27]  # Valores de inercia correspondientes

# Crear el gráfico
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para K-Means')
plt.xticks(k_values)  # Mostrar todas las etiquetas de k
plt.grid(True)

# Resaltar el "codo" (opcional, se puede ajustar visualmente)
plt.annotate('Codo', xy=(3, 60), xytext=(4, 70), arrowprops=dict(facecolor='black', shrink=0.05))

# Mostrar el gráfico
plt.show()
```

### Coeficiente de Silueta: Eligiendo el Número Óptimo de Clusters (k)

**Explicación:**
* Para cada punto de datos, se calcula su coeficiente de silueta, que mide qué tan similar es a su propio cluster en comparación con otros clusters.
* El coeficiente de silueta promedio para todos los puntos da una medida de qué tan bien se han agrupado los datos.
* Se pueden graficar los coeficientes de silueta para diferentes valores de *k* para ayudar a elegir el número óptimo de clusters (el *k* con el coeficiente de silueta promedio más alto suele ser preferible).

**Gráfico:** Un ejemplo de gráficos de silueta para diferentes valores de *k*, mostrando la distribución de los coeficientes de silueta para cada cluster.

---

```
# prompt: Un ejemplo de gráficos de silueta para diferentes valores de *k*, mostrando la distribución de los coeficientes de silueta para cada cluster.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

range_n_clusters = [2, 3, 4, 5]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init='auto')
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k" )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker="o",c="white", alpha=1, s=200, edgecolor="k")

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

### Métodos Avanzados de Clustering - Clustering Jerárquico

* **Idea Principal:** Construir una jerarquía de clusters.

* **Dos enfoques principales:**
    * **Aglomerativo (Bottom-up):** Comenzar con cada punto como un cluster individual y fusionar iterativamente los clusters más cercanos hasta que todos los puntos pertenezcan a un solo cluster o se alcance un criterio de parada.
    * **Divisivo (Top-down):** Comenzar con todos los puntos en un solo cluster y dividir iterativamente los clusters hasta que cada punto sea un cluster individual o se alcance un criterio de parada.

* **Dendrograma:** La jerarquía de clusters se representa mediante un dendrograma, un diagrama de árbol que muestra la secuencia de fusiones o divisiones. La altura de las ramas indica la distancia entre los clusters que se fusionaron o dividieron.

* **Ventajas:** No requiere especificar el número de clusters *a priori*. El dendrograma proporciona una visión rica de la estructura de los datos.

* **Desventajas:** Puede ser computacionalmente costoso para grandes conjuntos de datos. La interpretación del dendrograma puede ser subjetiva para determinar el número "correcto" de clusters.

**Diagrama:** Un ejemplo de un dendrograma mostrando la fusión jerárquica de clusters.

---

```
# prompt: Un ejemplo de un dendrograma mostrando la fusión jerárquica de clusters. Crea 100 datos aleatorios para este ejemplo

import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage

# Generar 100 datos aleatorios
np.random.seed(42)  # Para reproducibilidad
X = np.random.rand(100, 2)

# Calcular la matriz de enlace usando el método de Ward
linked = linkage(X, 'ward')

# Crear el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Dendrograma de Clustering Jerárquico')
plt.xlabel('Índice de los Datos')
plt.ylabel('Distancia')
plt.show()
```

### Métodos Avanzados de Clustering - DBSCAN

*DBSCAN (Density-Based Spatial Clustering of Applications with Noise)*

* **Idea Principal:** Agrupar puntos que están densamente empaquetados, marcando como ruido los puntos que se encuentran en regiones de baja densidad.

* **Parámetros clave:**
    * **`eps` (épsilon):** La distancia máxima entre dos puntos para que se consideren vecinos.
    * **`minPts`:** El número mínimo de puntos dentro del radio `eps` para que un punto central forme un cluster.

* **Tipos de puntos:**
    * **Punto Central:** Tiene al menos `minPts` vecinos dentro de `eps`.
    * **Punto Fronterizo:** No es un punto central, pero tiene al menos un punto central como vecino.
    * **Punto de Ruido:** No es un punto central ni un punto fronterizo.

* **Ventajas:** Puede descubrir clusters de formas arbitrarias. Robusto al ruido. No requiere especificar el número de clusters *a priori*.

* **Desventajas:** El rendimiento depende de la elección adecuada de `eps` y `minPts`, que puede ser difícil de determinar. Puede tener problemas con clusters de densidades muy variables.

**Diagrama:** Una ilustración que muestre puntos centrales, puntos fronterizos y puntos de ruido en un espacio bidimensional, formando clusters de formas no lineales.

---

```
# prompt: crea una dataset de 100 pares de números densos, para agruparlos usando DBSCAN. Luego crea una ilustración que muestre puntos centrales, puntos fronterizos y puntos de ruido en un espacio bidimensional, formando clusters de formas no lineales. crea una leyenda para los simbolos usados en el grafico.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Crea un dataset de 100 pares de números densos
np.random.seed(42)  # Para reproducibilidad
X = np.random.rand(100, 2)


# Aplica DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=10)  # Ajusta eps y min_samples según sea necesario
clusters = dbscan.fit_predict(X)

# Identifica los puntos centrales, fronterizos y de ruido
core_samples_mask = np.zeros_like(clusters, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
labels = dbscan.labels_

# Número de clusters encontrados
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

# Crea la gráfica
plt.figure(figsize=(8, 6))

# Colores para los diferentes tipos de puntos
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# Itera sobre los clusters
for k, col in zip(range(n_clusters_), colors):
    class_member_mask = (labels == k)
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=14, label=f'Punto Central - Cluster {k+1}' if k == 0 else None)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6, label=f'Punto Fronterizo - Cluster {k+1}' if k == 0 else None)


# Puntos de ruido
xy = X[labels == -1]
plt.plot(xy[:, 0], xy[:, 1], 'x', markeredgecolor='k', markersize=8, label='Punto de Ruido')

# Leyenda
plt.title(f'Clustering DBSCAN (eps={dbscan.eps}, min_samples={dbscan.min_samples})\n'
          f'Clusters: {n_clusters_}, Ruido: {n_noise_}')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()
plt.show()
```
